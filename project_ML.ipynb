# Preprocesamiento

```python
from sklearn.preprocessing import StandardScaler

scaler= StandardScaler()

dataframe["scaled_Amount"]= scaler.fit_transform(dataframe["Amount"].values.reshape(-1,1))

second_scaler=StandardScaler()

dataframe["scaled_Time"]= second_scaler.fit_transform(dataframe["Time"].values.reshape(-1,1))

dataframe.drop(["Amount", "Time"], axis=1, inplace=True)
```

# Separar entrenamiento y prueba

```python
from sklearn.model_selection import train_test_split

X= dataframe.drop("Class", axis=1)
y= dataframe["Class"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
```

```python
print("Proporción en y_train:")
print(y_train.value_counts(normalize=True))

print("\nProporción en y_test:")
print(y_test.value_counts(normalize=True))
```

# Primer Modelo : Regresión logística

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
model =  LogisticRegression(max_iter=100)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
cm =confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
plt.xlabel("predicción")
plt.ylabel("real")
plt.title("Matriz de Confusión - Regresión Logística")
plt.show()
```

# Modelo avanzado: RandomForest

```python
from sklearn.ensemble import RandomForestClassifier

rf= RandomForestClassifier(n_estimators=100, class_weight= "balanced", random_state=42)
rf.fit(X_train, y_train)
y_pred_rf= rf.predict(X_test)
print(classification_report(y_test, y_pred_rf))
```

```python
cm_rf= confusion_matrix(y_test, y_pred_rf)
sns.heatmap(cm_rf, annot=True, fmt="d", cmap="Greens")
plt.xlabel("Predicciones")
plt.ylabel("Reales")
plt.title("Matriz de Confusión - Random Forest")
plt.show()
```

```python
import numpy as np
feat_imp = rf.feature_importances_

indices = np.argsort(feat_imp)[::-1][:10]

plt.figure(figsize=(10, 6))
sns.barplot(x=feat_imp[indices], y=X.columns[indices])
plt.title("Top 10 Variables Más Importantes")
plt.xlabel("Importancia")
plt.ylabel("Variable")
plt.show()
```

```python
from sklearn.metrics import roc_auc_score, roc_curve, precision_recall_curve
import matplotlib.pyplot as plt

y_prob = rf.predict_proba(X_test)[:, 1] # [:, 1] significa solo probabilidad de que sea 1 
fpr, tpr, thresholds= roc_curve(y_test, y_prob) 
plt.figure(figsize=(6,4))
plt.plot(fpr, tpr, label="ROC Curve")
plt.plot([0, 1], [0, 1], 'k--', label="Azar")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate (Recall)")
plt.title("ROC Curve")
plt.legend()
plt.grid()
plt.show()
```

```python
roc_auc = roc_auc_score(y_test, y_prob)
print("AUC-ROC Score:", roc_auc)
```

```python
precision, recall, _ = precision_recall_curve(y_test, y_prob)

plt.figure(figsize=(6,4))
plt.plot(recall, precision, label="PR Curve")
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.title("Precision-Recall Curve")
plt.grid()
plt.legend()
plt.show()
```

# EXPLICABILIDAD DEL MODELO

```python
from sklearn.inspection import permutation_importance
import seaborn as sns
import matplotlib.pyplot as plt

results_pm= permutation_importance(rf, X_test, y_test, n_repeats=10, random_state=42, scoring="roc_auc")
sns.barplot(x=results_pm.importances_mean, y=X_test.columns)
plt.title("Importancia por permutación")
plt.xlabel("promedio de importancia")
plt.ylabel("Columna")
plt.grid()
plt.show()
```
